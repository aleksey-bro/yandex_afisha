### Описание данных

В распоряжении находится таблица заказов пользователей, объединённая с данными о мероприятиях и местах их проведения. Каждая строка таблицы соответствует одному заказу пользователя.

В таблице представлены следующие поля:

- `user_id` — уникальный идентификатор пользователя, совершившего заказ;
- `device_type_canonical` — тип устройства, с которого был оформлен заказ ( `mobile` — мобильные устройства, `desktop` — стационарные);
- `order_id` — уникальный идентификатор заказа;
- `order_dt` — дата создания заказа;
- `order_ts` — дата и время создания заказа;
- `currency_code` — валюта оплаты;
- `revenue` — выручка от заказа;
- `tickets_count` — количество купленных билетов;
- `days_since_prev` — количество дней от предыдущей покупки пользователя, для пользователей с одной покупкой — значение пропущено;
- `event_id` — уникальный идентификатор мероприятия;
- `service_name` — название билетного оператора;
- `event_type_main` — основной тип мероприятия (театральная постановка, концерт и так далее);
- `region_name` — название региона, в котором прошло мероприятие;
- `city_name` — название города, в котором прошло мероприятие.

Данные охватывают информацию о покупках пользователей, характеристиках мероприятий и географии их проведения, что позволяет анализировать пользовательскую активность, структуру продаж и региональные особенности спроса.

## 1. Загрузка данных и знакомство с ними

Начнём с загрузки библиотек и данных. Выгрузка будет производиться в датафрейм из базы данных PostgreSQL с помощью специальных библиотек. Также будем использовать pandas и библиотеки визуализации данных matplotlib и seaborn, а также phik_matrix для построения матрицы корреляции.

```python
#Загружаем библиотеки для подключения к базе данных
!pip install sqlalchemy
!pip install psycopg2-binary
!pip install phik
import sys
!{sys.executable} -m pip install python-dotenv
import psycopg2 as ps2

#Загружаем остальные библиотеки
import matplotlib.pyplot as plt
import seaborn as sns

#Также библиотека и функция для работы с датафреймами
import pandas as pd
from sqlalchemy import create_engine
from phik import phik_matrix
```

```python
from dotenv import load_dotenv
import os

# загрузка переменных из .env
load_dotenv()

#Записываем в объект данные для подключения к базе данных
db_config = {'user': os.getenv("DB_USER"), # имя пользователя
             'pwd': os.getenv("DB_PASSWORD"), # пароль
             'host': os.getenv("DB_HOST"),
             'port': os.getenv("DB_PORT"), # порт подключения
             'db': os.getenv("DB_NAME") # название базы данных
             }
#Передаем функции create_engine() строку для соединения с базой данных.
#Строка содержит информацию об используемой СУБД, а также все данные для авторизации
connection_string = 'postgresql://{}:{}@{}:{}/{}'.format(
    db_config['user'],
    db_config['pwd'],
    db_config['host'],
    db_config['port'],
    db_config['db'],
)
engine = create_engine(connection_string)

#Создаем SQL запрос для выгрузки данных 
query = '''
SELECT
    p.user_id,
    p.device_type_canonical,
    p.order_id,
    p.created_dt_msk AS order_dt,
    p.created_ts_msk AS order_ts,
    p.currency_code,
    p.revenue,
    p.tickets_count,
    (
        DATE(p.created_dt_msk)
        - LAG(DATE(p.created_dt_msk)) OVER (
            PARTITION BY p.user_id
            ORDER BY p.created_dt_msk
        )
    )::int AS days_since_prev,
    e.event_id,
    e.event_name_code AS event_name,
    e.event_type_main,
    p.service_name,
    r.region_name,
    c.city_name
FROM afisha.purchases p
INNER JOIN afisha.events e ON p.event_id = e.event_id
JOIN afisha.city c ON e.city_id = c.city_id
JOIN afisha.regions r ON c.region_id = r.region_id
WHERE p.device_type_canonical IN ('mobile', 'desktop') 
    AND e.event_type_main != 'фильм'
ORDER BY p.user_id
'''

#Результат выдачи SQL-запроса записываем в датафрейм с помощью метода pd.read_sql_query():
df = pd.read_sql_query(query, con=engine)
```

### Приведение выручки к единой валюте
Данные о выручке сервиса представлены в российских рублях и казахстанских тенге. Приведем выручку к единой валюте — российскому рублю.

Для этого используем датасет с информацией о курсе казахстанского тенге по отношению к российскому рублю за 2024 год — `final_tickets_tenge_df.csv`. 

Значения в рублях представлено для 100 тенге.

Результаты преобразования сохраним в новый столбец `revenue_rub`.

```python
# создаем копию датасета до преобразования для возможности проверить сделанные изменения после предобработки
temp = df.copy() 
len(temp)
```

```python
# Загрузка данных с курсом тенге
tenge_df = pd.read_csv('https://code.s3.yandex.net/datasets/final_tickets_tenge_df.csv')
tenge_df = tenge_df[['data', 'curs']]

tenge_df.head()
```

```python
tenge_df.info()
```

```python
# Приведем дату к формату datetime
tenge_df['data'] = pd.to_datetime(tenge_df['data'])
```

```python
# Объединяем два датасета по дате заказа
df = df.merge(
    tenge_df,
    left_on='order_dt',
    right_on='data',
    how='left'
)
df.info()
```

```python
# Проверяем наличие пропусков после объединения
print('\nКоличество пропусков в курсе после объединения:')
df['curs'].isna().sum()
```

Курс дан за 100 тенге, поэтому для перевода нужно:

revenue_rub = revenue*(kurs/100)

```python
# Создаем столбец revenue_rub
df['revenue_rub'] = df.apply(
    lambda row: row['revenue'] if row['currency_code'] == 'rub'
    else row['revenue'] * row['curs'] / 100,
    axis=1
)

#Удаляем ненужные столбцы
df = df.drop(columns=['data', 'curs'])
print('Проверка первых строк после конвертации:')

# Проверка для rub — значения должны совпадать
rub_check = df[df['currency_code'] == 'rub']
print('Разница revenue и revenue_rub для rub:')
print((rub_check['revenue'] - rub_check['revenue_rub']).abs().sum())

# Проверка для kzt — значения должны отличаться
kzt_check = df[df['currency_code'] == 'kzt']
print('\nПример конвертации для kzt:')
print(kzt_check[['revenue', 'revenue_rub']].head())
```

**Промежуточный вывод**

В исходных данных выручка представлена в двух валютах: rub и kzt.

Для заказов в тенге произведена конвертация в рубли с использованием официального курса за соответствующую дату заказа. 
Курс представлен в рублях за 100 тенге, поэтому при расчёте использована формула:

revenue_rub = revenue × (курс / 100)

Для заказов в рублях значения оставлены без изменений.

Проверка показала:
- для rub значения полностью совпадают;
- для kzt произведён корректный пересчёт;
- пропусков в курсе после объединения не обнаружено.

Результат сохранён в новый столбец `revenue_rub`.

## Создание профиля пользователя

В будущем отдел маркетинга планирует создать модель для прогнозирования возврата пользователей. Поэтому сейчас от нас требуется построить агрегированные признаки, описывающие поведение и профиль каждого пользователя.

### Задача 3.1

В рамках подготовки данных для будущей модели прогнозирования возврата пользователей
необходимо сформировать агрегированные признаки, описывающие поведение каждого клиента.

Для каждого пользователя будут рассчитаны:

- дату первого и последнего заказа;
- устройство, с которого был сделан первый заказ;
- регион, в котором был сделан первый заказ;
- билетного партнёра, к которому обращались при первом заказе;
- жанр первого посещённого мероприятия (используйте поле `event_type_main`);
- общее количество заказов;
- средняя выручка с одного заказа в рублях;
- среднее количество билетов в заказе;
- среднее время между заказами.

Также будут добавлены два бинарных признака:

- `is_two` — совершил ли пользователь 2 и более заказа;
- `is_five` — совершил ли пользователь 5 и более заказов.

Перед построением профиля данные будут отсортированы по времени заказа.

```python
# Сортируем данные по пользователю и времени заказа
df = df.sort_values(['user_id', 'order_ts'])

print("Данные отсортированы.")
```

```python
# Агрегация по пользователям
user_profile = df.groupby('user_id').agg(
    # Даты
    first_order_dt=('order_dt', 'first'),
    last_order_dt=('order_dt', 'last'),
    
    # Признаки первого заказа
    first_device=('device_type_canonical', 'first'),
    first_region=('region_name', 'first'),
    first_service=('service_name', 'first'),
    first_event_type=('event_type_main', 'first'),
    
    # Другие метрики
    total_orders=('order_id', 'nunique'),
    avg_revenue_rub=('revenue_rub', 'mean'),
    avg_tickets=('tickets_count', 'mean'),
    avg_days_between=('days_since_prev', 'mean')
).reset_index()

print("Размер пользовательского профиля:", user_profile.shape)
user_profile.head()
```

```python
# Добавляем бинарные признаки
user_profile['is_two'] = (user_profile['total_orders'] >= 2).astype(int)
user_profile['is_five'] = (user_profile['total_orders'] >= 5).astype(int)

print("Добавлены бинарные признаки.")
user_profile[['user_id', 'total_orders', 'is_two', 'is_five']].head()
```

### Промежуточный вывод

В результате построен агрегированный профиль пользователя.

Для каждого клиента рассчитаны:

- дата первого и последнего заказа;
- характеристики первого заказа (устройство, регион, партнёр, жанр);
- общее количество заказов;
- средняя выручка с заказа;
- среднее количество билетов;
- средний интервал между заказами;
- бинарные признаки активности (2+ и 5+ заказов).

Полученный датафрейм содержит по одной строке на пользователя и
может быть использован для дальнейшего моделирования возврата клиентов.

Структура профиля соответствует требованиям задачи и корректно агрегирует поведенческие признаки.

Следующий этап — исследование признаков, влияющих на возврат пользователей, то есть на совершение повторного заказа. Для этого используем профили пользователей.

### Исследование признаков первого заказа и их связи с возвращением

#### Задача 4.1.1

На данном этапе анализируются признаки первого заказа пользователя:

- тип первого мероприятия;
- тип устройства;
- регион проведения мероприятия;
- билетный оператор.

Цель — определить, равномерно ли распределены пользователи по сегментам
или существуют выраженные «точки входа» на платформу.

Напишем функцию для определения признаков первого заказа

```python
# Сохраним копию профиля пользователей после фильтрации
user_profile_temp = user_profile.copy()

def segment_distribution(df, column):
    segment_stats = (
        df
        .groupby(column)
        .agg(users_count=('user_id', 'count'))
        .sort_values(by='users_count', ascending=False)
    )
    
    segment_stats['share'] = (
        segment_stats['users_count'] / segment_stats['users_count'].sum()*100
    )
    
    return segment_stats
```

#### Задача 4.1.2

На данном этапе исследования необходимо проанализировать, влияют ли характеристики первого заказа пользователя на вероятность его возвращения на платформу.

Логика исследования основана на гипотезе о существовании «точек входа» — сегментов, через которые пользователи чаще становятся повторными клиентами. В качестве таких точек входа рассматриваются признаки первого заказа:

- тип первого мероприятия;
- устройство, с которого была совершена первая покупка;
- регион проведения мероприятия;
- билетный оператор, продавший билеты.

Для оценки возврата используются бинарные признаки:

- `is_two` — совершил ли пользователь два и более заказа;
- `is_five` — совершил ли пользователь пять и более заказов.

В рамках задачи анализируется распределение пользователей по сегментам и рассчитывается доля возврата в каждом из них. Полученные результаты позволяют определить, существуют ли сегменты с повышенной вероятностью повторной покупки и можно ли считать их успешными «точками входа» для дальнейших маркетинговых стратегий.

Для начала рассчитаем среднюю конверсию в повторный возврат

```python
overall_two = user_profile['is_two'].mean()
overall_five = user_profile['is_five'].mean()

print("Средний возврат (2+):", round(overall_two * 100, 2), "%")
print("Глубокая лояльность (5+):", round(overall_five * 100, 2), "%")
```

Напишем универсальную функцию для оценки доли пользователей, совершивших два заказа и более по каждому сегменту

```python
features = [
    'first_event_type',
    'first_device',
    'first_region',
    'first_service'
]

for feature in features:
    
    print(f"\n===== {feature} =====")
    
    result = (
        user_profile
        .groupby(feature)
        .agg(
            users=('user_id', 'count'),
            return_2=('is_two', 'mean'),
        )
        .reset_index()
        .sort_values('users', ascending=False)
    )
    
    # если регион или оператор — оставляем топ-10
    if feature in ['first_region', 'first_service']:
        result = result.head(10)
        
    result['return_2_%'] = result['return_2'] * 100
    
    display(result)
```

Напишем универсальную функцию для визуализации распределений

```python
def plot_return_rate(data, segment_col, metric_col, title, top_n=10):
    baseline = user_profile['is_two'].mean()*100
    result = (
        data
        .groupby(segment_col)
        .agg(
            users=('user_id', 'count'),
            return_rate=(metric_col, 'mean')
        )
        .reset_index()
    )
    
    result['return_rate_%'] = result['return_rate'] * 100
    
    # сортировка по доле возврата
    result = result.sort_values('return_rate_%', ascending=False)
    
    # если сегментов много — берём топ-N по количеству пользователей
    if len(result) > top_n:
        result = result.sort_values('users', ascending=False).head(top_n)
        result = result.sort_values('return_rate_%', ascending=False)
    
    plt.figure(figsize=(8, 6))
    
    plt.bar(result[segment_col], result['return_rate_%'])
    
    # Добавляем линию со средним по выборке
    plt.axhline(
    y=baseline,
    linestyle='--'
    )
    plt.text(
        0,
        baseline,
        f'Среднее: {baseline:.1f}%',
        va='bottom'
    )
    
    plt.xticks(rotation=45, ha='right')
    plt.xlabel('Доля пользователей с 2+ заказами (%)')
    plt.title(title)
    plt.show()
```

Далее поочередно построим визуализацию для каждого признака

```python
plot_return_rate(
    user_profile,
    'first_event_type',
    'is_two',
    'Возврат (2+) по типу первого мероприятия'
)
```

```python
plot_return_rate(
    user_profile,
    'first_device',
    'is_two',
    'Возврат (2+) по устройству'
)
```

```python
plot_return_rate(
    user_profile,
    'first_region',
    'is_two',
    'Возврат (2+) по региону',
    top_n=10
)
```

```python
plot_return_rate(
    user_profile,
    'first_service',
    'is_two',
    'Возврат (2+) по билетному оператору',
    top_n=10
)
```

Наблюдается выраженная концентрация значений около целых чисел (2, 3, 4), 
что объясняется дискретной природой данных.

Сегменты с 4–5 билетами представлены значительно меньшим числом пользователей, 
что указывает на наличие правого хвоста распределения.

Таким образом, пользователи преимущественно совершают заказы небольшого размера, 
а крупные средние значения количества билетов встречаются существенно реже.

```python
# создаём категории
def ticket_segment(x):
    if 1 <= x < 2:
        return '1–2'
    elif 2 <= x < 3:
        return '2–3'
    elif 3 <= x < 5:
        return '3–5'
    elif x >= 5:
        return '5+'
    else:
        return 'Other'

user_profile['ticket_segment'] = user_profile['avg_tickets'].apply(ticket_segment)
user_profile.head()
```

```python
# Подсчитаем пользователей и долю возврата 
segment_stats = (
    user_profile
    .groupby('ticket_segment')
    .agg(
        users_count=('user_id', 'count'),
        return_rate=('is_two', 'mean')
    )
    .sort_index()
)

segment_stats['return_rate'] = segment_stats['return_rate'] * 100

segment_stats
```

```python
plt.figure(figsize=(10, 6))

plt.bar(segment_stats.index, segment_stats['return_rate'])

plt.xlabel('Сегмент по среднему количеству билетов')
plt.ylabel('Доля повторных покупок (%)')
plt.title('Возврат пользователей по сегментам среднего количества билетов')

plt.xticks(rotation=45)
plt.show()
```

- Проанализируем полученные результаты. Если полученные значения будут близки к нулю, проверим разброс данных в `total_orders`. Такое возможно, когда в данных преобладает одно значение: в таком случае корреляционный анализ может показать отсутствие связей. Чтобы этого избежать, выделим сегменты пользователей по полю `total_orders`, а затем повторим корреляционный анализ. Выделим такие сегменты:
    - 1 заказ;
    - от 2 до 4 заказов;
    - от 5 и выше.
- Визуализируем результат корреляции с помощью тепловой карты.

Для начала вернемся к копии датафрейма до добавления вспомогательных столбцов

```python
user_profile = user_profile_temp.copy()
```

```python
# Вычисляем корреляционную матрицу с использованием phi_k
# Исключим из анализа столбцы user_id (это просто идентификатор пользователя), is_two и is_five (так как это солбцы, связанные с количеством заказов)
df_corr = user_profile.drop(columns=['user_id', 'is_two', 'is_five'])
сorrelation_matrix = df_corr.phik_matrix(interval_cols=['total_orders', 'avg_revenue_rub', 'avg_tickets', 'avg_days_between'])

# Выводим результат
print('Корреляционная матрица с коэффициентом phi_k для user_profile')
сorrelation_matrix
```

```python
# Строим тепловую карту
plt.figure(figsize=(2, 6))

# Сохраняем матрицу корреляции количества заказов с другими признаками пользователя
data_heatmap = сorrelation_matrix.loc[сorrelation_matrix.index != 'total_orders'][['total_orders']].sort_values(by='total_orders', ascending=False)
sns.heatmap(data_heatmap,
            annot=True, # Отображаем численные значения в ячейках карты
            fmt='.2f', # Форматируем значения корреляции: два знака после точки
            cmap='coolwarm', # Устанавливаем цветовую гамму от красного (макс. значение) к синему
            linewidths=0.5, # Форматируем линию между ячейками карты
            cbar=False # Отключаем цветовую шкалу
           )

# Добавляем заголовок и подпись по оси Х
plt.title('Тепловая карта коэффициента phi_k \n для данных total_orders')
plt.xlabel('Количество заказов')

plt.yticks(rotation=360)
# Выводим график
plt.show()
```

2. Дата последнего заказа — **0.41**
3. Средний интервал между покупками — **0.36**
4. Среднее количество билетов — **0.34**
5. Средняя выручка — **~0.25**

Наименее связаны:

- регион,
- устройство,
- билетный оператор,
- жанр первого события.

---

### Итоговый вывод

Основной особенностью базы является высокая доля пользователей с одним заказом при среднем lifetime около **78 дней**.

Количество заказов определяется прежде всего:

- частотой покупок,
- интервалом между ними,
- объёмом заказа,

а не характеристиками первого взаимодействия.

Сокращение среднего интервала между заказами с ~19 дней до ~10 дней
характерно для пользователей с 5+ покупками.

### Рекомендации заказчику

#### Фокус на вторую покупку

Основная зона роста — пользователи с 1 заказом.

Рекомендуется:
- запускать триггерные коммуникации в первые 7–14 дней после покупки;
- использовать персонализированные рекомендации;
- предлагать скидку или бонус на вторую покупку.

---

#### Контроль интервала между покупками

Интервал — один из сильнейших факторов.

Рекомендуется:
- отслеживать пользователей с растущим интервалом,
- запускать реактивационные кампании,
- внедрить модель прогнозирования оттока.

---

#### Развитие сегмента 5+ заказов

Активные пользователи:
- имеют в 2 раза меньший интервал между покупками,
- формируют стабильную выручку.

Рекомендуется:
- запускать программу лояльности,
- предоставлять ранний доступ к событиям,
- использовать VIP-коммуникации.

---

#### Не делать ставку на статические признаки

Регион, устройство, жанр первого события показывают слабую связь с активностью.

Маркетинговая стратегия должна строиться на динамике поведения, а не на атрибутах первого заказа.
